```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("E:/研究生/好好学习之各类课程/Datawhale_Learning/CDNOW.txt",sep = '\s+', names=['用户ID', '购买日期','购买数量', '购买金额'])
data.head()
```

```python
data.describe()
```

### 针对于购买日期

```python
data1 = data.groupby('购买日期').用户ID .count()
# 可以看出某天的总购买量
data1
```

```python
data.groupby('购买日期').用户ID .count().sort_values(ascending = False)
# 按购买总数量降序排列
```

```python
plt.plot(data1);
```

```python
data.groupby('购买日期').购买金额.agg(['max','min'])
```
